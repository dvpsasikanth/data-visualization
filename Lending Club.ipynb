---

converting object datatypes to int

```python
loan_term = data['term'].str.extract('(\d+)').astype(int)
interest_rate = data['int_rate'].str.extract('(\d+)').astype(int)
emp_tenure = data['emp_length'].str.extract('(\d+)').astype(int)
```

```python
data['loan_term'] = loan_term
data['interest_rate'] = interest_rate
data['emp_tenure'] = emp_tenure
data.head()
```

```python
data.emp_title.value_counts().nlargest(50).plot(kind='bar',figsize=(10,5))
plt.title("Loan count based on Employment")
plt.ylabel('loan count')
plt.xlabel('emp_title')
```

```python
data.home_ownership.value_counts().plot(kind='bar',figsize=(10,5))
plt.title("Loan count based on home ownership")
plt.ylabel('loan count')
plt.xlabel('home_ownership')
```

```python
people_count = data.groupby(['addr_state'])['funded_amnt'].count().nlargest(7).reset_index(name='Count of people')
people_count
```

```python
state_count_arrray = people_count[["addr_state"]].as_matrix().ravel()
People_array = people_count[["Count of people"]].as_matrix().ravel()

x = np.arange(people_count.shape[0])
plt.bar(x, People_array)
plt.xticks(x, state_count_arrray)
plt.ylabel("Number of People")
```

```python
plt.pie(People_array, labels=state_count_arrray)
```

```python
data.title.value_counts().nlargest(15).plot(kind='bar',figsize=(10,5))
plt.title("Loan count based on the purpose of the loan")
plt.ylabel('loan count')
plt.xlabel('purpose of the loan')
```
