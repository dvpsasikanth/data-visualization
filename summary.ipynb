### dropping the column str_nm as we don't use it in the model

```python
finaldata1 = data1.drop(['str_nm'],axis=1)
```

### dropping the column rtl_mo_nbr,rtl_mo_strt_dt as we don't use in the model

```python
finaldata2 = data2.drop(columns = ['rtl_mo_nbr','rtl_mo_strt_dt'],axis=1)
```

### joining  category_sales and store_lookup table on str_id field

```python
result = pd.merge(data,finaldata1[['str_id', 'channel_rollup']],on='str_id')
```

### joining category_sales and date-dim tables on month_key field

```python
result1 = pd.merge(result,
                 finaldata2[['month_key', 'rtl_yr']],
                 on='month_key')
```

```python
result1.head(10)
```

### storing result1 variable in the name of finalresult

```python
finalresult = result1
```

```python
finalresult.head(20)
```

```python
finalresult.tail(20)
```

### replacing null values in running_sales column using aggregate function mean

```python
finalresult['running_sales'] = finalresult['running_sales'].fillna(finalresult.groupby('str_id')['running_sales'].transform('mean'))
```

### aggregate functions on running_sales column by applying groupby on channel_rollup,rtl_yr

```python
output = finalresult.groupby(['channel_rollup','rtl_yr']).agg(avg_running_sales=('running_sales', 'mean'), 
                               total_runningsales=('running_sales', 'sum'))
```
